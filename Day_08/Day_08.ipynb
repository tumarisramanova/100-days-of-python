```python
def life_in_weeks(age):
    weeks_left = (90 - age) * 52
    print(f"You have {weeks_left} weeks left.")

life_in_weeks(22)
```

```python
def calculate_love_score(name1, name2):
    # Convert both names to lowercase to make the counting case-insensitive
    combined_names = (name1 + name2).lower()

    # Count letters for TRUE
    t = combined_names.count("t")
    r = combined_names.count("r")
    u = combined_names.count("u")
    e = combined_names.count("e")
    true_score = t + r + u + e

    # Count letters for LOVE
    l = combined_names.count("l")
    o = combined_names.count("o")
    v = combined_names.count("v")
    e = combined_names.count("e")
    love_score = l + o + v + e

    # Combine scores
    total_score = int(str(true_score) + str(love_score))

    print(total_score)

# Test example
calculate_love_score("Kanye West", "Kim Kardashian")
```

```python
# Caesar Cipher
alphabet = [chr(i) for i in range(97, 123)]

def caesar(start_text, shift_amount, cipher_direction):
    end_text = ""
    if cipher_direction == "decode":
        shift_amount *= -1
    for letter in start_text:
        if letter in alphabet:
            position = alphabet.index(letter)
            new_position = (position + shift_amount) % 26
            end_text += alphabet[new_position]
        else:
            # Keep spaces, numbers, punctuation unchanged
            end_text += letter
    print(f"Here's the {cipher_direction}d result: {end_text}\n")

# Repeat until user wants to stop
should_continue = True
while should_continue:
    direction = input("Type 'encode' to encrypt, type 'decode' to decrypt:\n").lower()
    text = input("Enter your message:\n").lower()
    shift = int(input("Enter shift number:\n")) % 26  # prevent large numbers

    caesar(start_text=text, shift_amount=shift, cipher_direction=direction)

    restart = input("Type 'yes' if you want to go again, otherwise type 'no':\n").lower()
    if restart != "yes":
        should_continue = False
        print("Goodbye")


```
