```python
import random

choose = int(input("What do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors: "))

if choose == 0:
    print("You chose Rock")
elif choose == 1:
    print("You chose Paper")
elif choose == 2:
    print("You chose Scissors")
else:
    print("Wrong choice!")
    exit()

computer = random.randint(0, 2)
print("Computer chose:", computer)

# Decide winner
if computer == 0 and choose == 1:
    print("You win")
elif computer == 1 and choose == 0:
    print("You lose")
elif computer == 1 and choose == 2:
    print("You win")
elif computer == 2 and choose == 1:
    print("You lose")
elif computer == 0 and choose == 2:
    print("You lose")
elif computer == 2 and choose == 0:
    print("You win")
elif computer == choose:
    print("It's a draw")
else:
    print("Error in choices")

```

```python
import random

# ASCII Art
rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''

options = [rock, paper, scissors]

# User choice
choose = int(input("What do you choose? Type 0 for Rock, 1 for Paper or 2 for Scissors: "))

if choose >= 0 and choose <= 2:
    print("You chose:")
    print(options[choose])
else:
    print("Wrong choice!")
    exit()

# Computer choice
computer = random.randint(0, 2)
print("Computer chose:")
print(options[computer])

# Decide winner
if computer == choose:
    print("It's a draw ")
elif (choose == 0 and computer == 2) or (choose == 1 and computer == 0) or (choose == 2 and computer == 1):
    print("You win")
else:
    print("You lose ")
```
